```python
import json
import os
from os.path import realpath, dirname
ROOT_PATH = dirname(realpath(os.getcwd()))
DATA_PATH = os.path.join(ROOT_PATH, 'data','preprocessed')
with open(os.path.join(DATA_PATH, 'preprocessed_data.json'),'r') as f:
    data = json.load(f)
```

```python
import requests
import numpy as np
def load_image(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()  # Check if the request was successful
        image_array = np.frombuffer(response.content, np.uint8)
        return image_array
    except requests.exceptions.RequestException as e:
        print(f"Error fetching image from {url}: {e}")
        return None

for d in data[:5]:
    image_urls = d.get('image')
    # print(image_url)
    for image_url in image_urls:
        image_array = load_image(image_url)
        if image_array is None:
            continue
        print(f"Loaded image of shape: {image_array.shape} from URL: {image_url}")
```
