```python
import json
import os
import gzip

data_path = 'Clothing_Shoes_and_Jewelry_5.json.gz'

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def get_data(dp=data_path):
    data = []
    for d in parse(dp):
        data.append(d)
    return data

data = get_data()
print(data[0])
```
